## Exercise 2 - Conduct the Logistic Regression

Now it's time to conduct the logistic regression.

```python
#import the statsmodels
import statsmodels.formula.api as smf
```

Rename the columns

```python
ad_data = ad_data.rename(columns={'Clicked on Ad': 'CoA',
                                      'Daily Time Spent on Site': 'DTSoS',
                                      'Area Income': 'AI',
                                      'Daily Internet Usage': 'DIU'})
```

Specify the model

```python
specification = 'CoA ~ DTSoS + Age + AI + DIU + Male'
```

Conduct the logistic regression

```python
mod = smf.logit(specification, data=ad_data)
res = mod.fit()
print(res.summary())
```

## Extra

Now it's time to do a train test split, and train our model for prediction tasks!

You'll have the freedom here to choose columns that you want to train on!

```python
ad_data = pd.read_csv('Data/10.1 advertising.csv')
```

Split the data into training set and testing set using train_test_split

```python
from sklearn.model_selection import train_test_split

X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = ad_data['Clicked on Ad']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

Train and fit a logistic regression model on the training set.

```python
from sklearn.linear_model import LogisticRegression

logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

Now predict values for the testing data.

```python
predictions = logmodel.predict(X_test)
```

Create a classification report for the model.

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predictions))
```

Create a confusion matrix for the model.

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, predictions)
```

Draw ROC-curve for the model and compute AUC.

```python
from sklearn.metrics import roc_curve, auc

y_predict_probabilities = logmodel.predict_proba(X_test)[:,1]

fpr, tpr, thr = roc_curve(y_test, y_predict_probabilities)

#print(fpr)
#print(tpr)
#print(thr)

roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange',
         lw=2, label='ROC curve (area = %0.2f)' % roc_auc)

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
