# Matplotlib

Advantages:
* easy to get started for simple plots
* support for custom labels and texts
* great control of every element in a figure
* high-quality output in many formats
* very customizable in general

**Canvas**: area on which plot(s) should be displayed
* `plt.subplot(nrows, ncols, plot_number)`
  * first two arguments stay fixed for one canvas
  * last argument accesses the specific plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
# generate data to plot
x = np.linspace(0,10, 20)
y = x**2

# create a plot
plt.plot(x,y, 'orange')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('plot title')
plt.show()
```

```python
# creates a subplot with one row and 2 columns
# accessing first plot
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')

# accessing second plot
plt.subplot(1,2,2)
plt.plot(y,x, 'g*-')
```

```python
# creates a subplot with 2 rows and 1 column
# accessing first plot
plt.subplot(2,1,1)
plt.plot(x,y, 'r--')

# accessing second plot
plt.subplot(2,1,2)
plt.plot(y,x, 'g*-')
```

```python
# creates a subplot with 2 rows and 2 column -> space for 4 plots
# accessing first plot
plt.subplot(2,2,1)
plt.plot(x,y, 'r--')

# accessing second plot
plt.subplot(2,2,2)
plt.plot(y,x, 'g*-')
```

```python
# creates a subplot with 2 rows and 2 column -> space for 4 plots
# accessing first plot
plt.subplot(2,2,1)
plt.plot(x,y, 'r--')

# accessing second plot
plt.subplot(2,2,2)
plt.plot(y,x, 'g*-')

# accessing third plot
plt.subplot(2,2,3)
plt.plot(x,y, 'r--')

# accessing fourth plot
plt.subplot(2,2,4)
plt.plot(y,x, 'g*-')

```

## Object Oriented Method

* to create figure objects and then just call methods or attributes off of that object
* nicer when dealing with a canvas that has multiple plots on it
* **always use axes object to add stuff to the plot**
  * after creating the axes, the figure is not called again, only the axes object

<br> 

| method | use |
|--------|-----|
| `.figure()`| creates empty canvas |
| `.plot(x, y, arguments)` | specifies what should be plotted <br> e.g. `label=`, to add labels 
| `.add_axes([left, bottom, width, height])`| adds a set of axes <br> `left`: position of the left edge of the axes relative to the left edge of the figure <br> `bottom`: position of the bottom edge of the axes relative to the bottom edge of the figure <br> `width`: width of the axes relative to the width of the figure <br> `height`: height of the axes relative to the height of the figure|
| `.set_xlabel()` | x-axis label |
| `.set_ylabel()`| y-axis label |
| `.set_title()`| plot title |
| `plt.tight_layout()` | avoid overlap in subplots|
| `.savefig('file_name', dpi=)`| to PNG, JPG, EPS, SVG, PGF and PDF <br> even with specified `dpi`possible |
| `.legend()` | adds a legend <br> **Caution**: only possible to add if in `.plot()` `label=` has been specified!|
| `.set_ylim` | custom y axis range |
| `.set_xlim` | custom x axis range |
| `.axis('tight')` | automatically tightly fitted axes ranges |

<br> 
<br>

| keyword arguments | method | use |
|--------|-----| -----|
| `figsize=` |`.figure()` <br> `.subplots()`|tuple of width and height of the figure in *inches* |
| `dpi=` | `.savefig()` <br> `.figure()` <br> `.subplots()`| pixel per inch |
| `label=` | `.plot()` | adds labels under |
| `loc=` | `.legend()` | specifies *location* of legend, e.g `'lower right'`, `'best'` |
| `color=` | `.plot()` | specifies the color: hexcode, color name, letter |
| `alpha=` | `.plot()` | indicating opacity |
| `linewidth=`, `lw=` | `.plot()` | change the line width |
| `linestyle=`, `ls=` | `.plot()` | change the line style, e.g.: `-`, `.-`, `:` <br> also possible to build custom dashes *(see below)* |
| `marker=`| `.plot()` | change the marker, e.g.: `+`, `o`, `1` |
| `markersize=`| `.plot()` | change the marker size |
| `markerfacecolor=`| `.plot()` | change the marker color |
| `markeredgewidth=`| `.plot()` | change the marker edge |
| `markeredgecolor=`| `.plot()` | change the marker edge color |

```python
# create a figure i.e. an empty canvas
fig = plt.figure()

# add axes to the figure
axes = fig.add_axes([90, 0.1, 0.8, 0.5])

# plot on axes set
axes.plot(x,y, 'purple')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
# create a figure i.e. an empty canvas
fig_multi = plt.figure()

# create multiple axes
axes1 = fig_multi.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig_multi.add_axes([0.4, 0.4, 0.6, 0.3]) # insert axes

# larger figure axes 1
axes1.plot(x, y, 'orange')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# insert figure axes 2
axes2.plot(y, x, 'purple')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
```

```python
fig, axes = plt.subplots()

axes.plot(x,y, 'purple')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig, axes = plt.subplots(2, 2)

# to access specific plot in the canvas use indexing
axes[0,1].plot(x, y, 'blue')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')
axes[0,1].set_title('title')

axes[1,0].plot(x, y, 'purple')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')
axes[1,0].set_title('title')
```

```python
fig, axes = plt.subplots(1, 2)

for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,4), dpi=70)  # doesn't show anything, it's just an object
axes = fig.add_axes([0, 0, 1, 1])        # shows empty canvas
axes.plot(x, y, 'r')                     # shows the plot
```

```python
fig, axes = plt.subplots(figsize=(12,3), dpi=70)

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend(loc='lower right')
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.1) # half-transparant, alpha determines opacity
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")  
```

```python
ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)
fig
```

```python
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')
fig
```

```python
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...
fig
```

```python
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')
fig
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
